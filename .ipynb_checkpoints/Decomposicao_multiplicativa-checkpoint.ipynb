```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import  matplotlib.pyplot  as  plt
```

```python
df  =  pd.read_csv('airline-passengers.csv', header = 0, index_col = 0, parse_dates = True, squeeze = True)
```

```python
df
```

```python
df.plot()
plt.ylabel('passageiros')
plt.title("Passageiros de companhias aereas")
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import  statsmodels.api as sm
```

```python
decompTurnover_df = sm.tsa.seasonal_decompose(df, model="additive")
decompTurnover_df.plot()
plt.show()
```
